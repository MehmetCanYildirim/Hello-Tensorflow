# Çizgi Grafiği

```
x = np.linspace(0, 20, 1000)
```

```
y = np.sin(x) + 0.2 * x
```

```
plt.plot(x, y)
plt.show()
```

```
plt.plot(x, y)
plt.xlabel('girdi')
plt.ylabel('çıktı')
plt.title('çizgi grafiği')
```

# Serpilme Grafiği

```
X = np.random.randn(100, 2)
np.mean(X)
```

```
plt.scatter(X[:, 0], X[:, 1])
```

```
X = np.random.randn(200, 2)
print("Tüm noktaların aritmetik ortalaması: ", np.mean(X))

X[:50] += 3
print("0 ve 50 arasındaki noktaların aritmetik ortalaması: ", np.mean(X[:50]))
print("50 ve 200 arasındaki noktaların aritmetik ortalaması: ", np.mean(X[50:200]))
print("Değiştirilmiş değerlerle tüm noktaların aritmetik ortalaması: ", np.mean(X[:200]))

Y = np.zeros(200)
Y[:50] = 1
plt.scatter(X[:, 0], X[:, 1], c=Y)
```
